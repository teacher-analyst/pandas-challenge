```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("schools_complete.csv")
student_data_to_load = Path("students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_students = school_data_complete["Student ID"].count()
total_schools = school_data["School ID"].count()
total_budget = school_data["budget"].sum()
avg_mathscore = student_data["maths_score"].mean()
avg_readingscore = student_data["reading_score"].mean()
maths_pass_percent = len(student_data[student_data["maths_score"]>=50])/total_students*100
reading_pass_percent = len(student_data[student_data["reading_score"]>=50])/total_students*100
overall_passing_percent= len(student_data[(student_data["maths_score"]>=50) & (student_data["reading_score"]>=50)])/total_students*100
```

```python
area_summary_df =pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Maths Score": [avg_mathscore],
    "Average Reading Score": [avg_readingscore],
    "% Passing Maths": [maths_pass_percent],
    "% Passing Reading": [reading_pass_percent],
    "% Overall Passing": [overall_passing_percent]   
})

area_summary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Created a dataframe and modified it
summary_df = school_data

#set school name as the index
df = summary_df.set_index("school_name")

#deleted "School ID" column
del df["School ID"]

#Renamed columns and calculated "Total School Budget
school_summary_df = df.rename(columns={"type": "School Type", "size": "Total Students",
                                              "budget": "Total School Budget"})

#Calculated "Per Student Budget"
school_summary_df["Per Student Budget"]=school_summary_df["Total School Budget"]/school_summary_df["Total Students"]

#Calculate average maths and reading scores
merged_df = school_data_complete 
avg_math_score= merged_df.groupby("school_name").mean()["maths_score"]
avg_reading_score = merged_df.groupby("school_name").mean()["reading_score"]

```

```python
#Calculate the % passing maths and % passing reading
per_school_count = school_data_complete["school_name"].value_counts()
passing_math = school_data_complete[school_data_complete["maths_score"]>=50]
passing_reading = school_data_complete[school_data_complete["reading_score"]>=50]

passing_math_percentage = passing_math.groupby(["school_name"]).count()["student_name"]/per_school_count*100
passing_reading_percentage = passing_reading.groupby(["school_name"]).count()["student_name"]/per_school_count*100

```

```python
#Calculate overall passing
overall_passing = school_data_complete[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]

overall_passing_percentage = overall_passing.groupby(["school_name"]).count()["student_name"]/per_school_count*100

overall_passing_percentage
```

```python
#Dataframe to hold the relevant results
school_summary_df['Average Maths Score'] = avg_math_score
school_summary_df['Average Reading Score'] = avg_reading_score
school_summary_df['% Passing Maths'] = passing_math_percentage
school_summary_df['% Passing Reading'] = passing_reading_percentage
school_summary_df['% Overall Passing'] = overall_passing_percentage
school_summary_df
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_schools_df = school_summary_df.sort_values(["% Overall Passing"], ascending=False)
top_schools_df.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
bottom_schools_df = school_summary_df.sort_values(["% Overall Passing"], ascending=True)
bottom_schools_df.head(5)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#filtering the data by year levels
year9= school_data_complete.loc[(school_data_complete ["year"] == 9)]
year10 = school_data_complete.loc[school_data_complete ["year"] == 10]
year11 = school_data_complete.loc[school_data_complete ["year"] == 11]
year12 = school_data_complete.loc[school_data_complete ["year"] == 12]

#mean maths score for each year level by school 
year9_maths_score = year9.groupby(["school_name"]).mean()["maths_score"]
year10_maths_score = year10.groupby(["school_name"]).mean()["maths_score"]
year11_maths_score = year11.groupby(["school_name"]).mean()["maths_score"]
year12_maths_score = year12.groupby(["school_name"]).mean()["maths_score"]

```

```python
#Combine average maths scores in a dataframe

maths_scores_by_year=pd.DataFrame(year9_maths_score)
maths_scores_by_year['Year 10'] = year10_maths_score
maths_scores_by_year['Year 11'] = year11_maths_score
maths_scores_by_year['Year 12'] = year12_maths_score
maths_scores_by_year= maths_scores_by_year.rename(columns={'maths_score': 'Year 9'})
maths_scores_by_year
```

## Reading Score by Year

* Perform the same operations as above for reading scores

```python
#Note: I did not repeat the loc filter here as those were already performed above before average math scores were calculated. 

#mean reading score for each year level by school 
year9_reading_score = year9.groupby(["school_name"]).mean()["reading_score"]
year10_reading_score = year10.groupby(["school_name"]).mean()["reading_score"]
year11_reading_score = year11.groupby(["school_name"]).mean()["reading_score"]
year12_reading_score = year12.groupby(["school_name"]).mean()["reading_score"]
```

```python
#Combine average reading scores in a dataframe

reading_scores_by_year=pd.DataFrame(year9_reading_score)
reading_scores_by_year['Year 10'] = year10_reading_score
reading_scores_by_year['Year 11'] = year11_reading_score
reading_scores_by_year['Year 12'] = year12_reading_score
reading_scores_by_year = reading_scores_by_year.rename(columns={'reading_score': 'Year 9'})
reading_scores_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Created 4 bins 
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

 # Slice the data and place it into bins
pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels= spending_labels).head()

 # Place the data series into a new column inside of the DataFrame
school_summary_df["Spending Ranges(Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels= spending_labels)
school_summary_df
```

```python
#Scores by school spending
spending_summary_df= school_summary_df.groupby("Spending Ranges(Per Student)")

spending_summary_df[["Average Maths Score", "Average Reading Score", "% Passing Maths", "% Passing Reading", "% Overall Passing"]].mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Created 4 bins 
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Used pd.cut on the "Total Students" column of the school_summary_df in 
#the part titled School Summary
pd.cut(school_summary_df["Total Students"], size_bins, labels= size_labels)

 # Place the data series into a new column inside of the school_summary_df DataFrame
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], size_bins, labels= size_labels)
school_summary_df
```

```python
#Scores by school size
size_summary_df= school_summary_df.groupby("School Size")

size_summary_df[["Average Maths Score", "Average Reading Score", "% Passing Maths", "% Passing Reading", "% Overall Passing"]].mean()
```
